```python
import cv2
import matplotlib.pyplot as plt
import os

raw_path = "/content/drive/MyDrive/CSC2427_SC549_PA_03/Extract_images/images_1"
frames = sorted(os.listdir(raw_path))[:10]

plt.figure(figsize=(12,6))
for i, f in enumerate(frames):
    img = cv2.imread(os.path.join(raw_path, f))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(2,5,i+1)
    plt.imshow(img)
    plt.title(f)
    plt.axis("off")
plt.show()
```
